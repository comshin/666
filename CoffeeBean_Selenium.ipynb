```python
# 라이브러리
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from bs4 import BeautifulSoup
import pandas as pd

# 웹드라이버 지정
wd = webdriver.Chrome('./WebDriver/chromedriver.exe')

# 메인코드
result = []
for i in range(1, 400):
    try:
        wd.get('https://www.coffeebeankorea.com/store/store.asp')
        wd.execute_script('storePop2(%d)' %i)
        # execute_script : js코드를 인자로 넣을 수 있어 브라우저 화면을 제어
        element = WebDriverWait(wd, 2).until(
            expected_conditions.presence_of_element_located((By.CLASS_NAME, "store_table")))
        # WebDriverWait(wd, 2) : WebDriverWait을 통해 wd를 "최대" 2초동안 기다린다. 
        # expected_conditions: 조건을 설정하고 조건에 부합하는지 판단
        # .until(EC.presence_of_element_located((By.ID, "myDynamicElement"))
        # 언제까지? -> ID가 "myDynamicElement"인 엘리먼트가 나올 때까지
        # (이때 해당 엘리먼트가 나오면, EC에서 True를 리턴한다.)
        
        html = wd.page_source
        soup = BeautifulSoup(html, 'html.parser')
        store_name = soup.select('.store_txt > h2')[0].string
        store_info = soup.select('table.store_table > tbody > tr > td')
        store_address = list(store_info[2])[0]
        store_phone = store_info[3].string
        result.append([store_name, store_address,  store_phone])
        print(f'{i} : {store_name} {store_address} {store_phone}')
    except:
        print(f'{i} : not exist')
        continue
wd.quit()
df = pd.DataFrame(result, columns = ('name', 'address', 'phone'))
df.to_csv('CoffeeBean3.csv', encoding='utf-8', mode='w', index=False)
# CoffeeBean3.csv파일로 저장
print('Completed..')
```
