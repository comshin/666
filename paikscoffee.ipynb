```python
# 전국 빽다방 매장 정보 수집
```

```python
import time
import requests
import pandas as pd
```

```python
page_num = 1
page_list = []

while True:
    url = f"https://theborndb.theborn.co.kr/wp-json/api/get_store/?state=9&category=275&depth1=&depth2=&paged={page_num}"

    response = requests.get(url)
    infomation = response.json()['results']

    # 마지막 페이지를 넘어간 경우
    if infomation is None:
        print('모든 매장 수집 완료')
        break

    page_list.extend(infomation)
    print(f'{page_num}페이지 수집 완료')
    page_num += 1
    time.sleep(0.01)
```

```python
df_all = pd.DataFrame(page_list)
```

```python
# 20230220 기준 1265개 매장 정보
df_all
```

```python
# 필요한 컬럼만 선택
paiks_coffee = df_all.loc[:, ['store_brand', 'store_name', 'store_depth1', 'store_address']]
```

```python
paiks_coffee
```

```python
# 컬럼 이름 한글로 변경
paiks_coffee = paiks_coffee.rename(columns={'store_brand':'브랜드명',
                                            'store_name':'지점명',
                                            'store_depth1':'시도명',
                                            'store_address':'주소'})
```

```python
paiks_coffee
```

```python
# 지점명에 중복이 있다?
paiks_coffee.nunique()
```

```python
# 같은 지점명 2개가 2개씩 들어있다
paiks_coffee.loc[paiks_coffee.duplicated(['지점명'], keep=False), :]
```

```python
# 지점명이 같은 행은 하나만 두고 제거
paiks_coffee = paiks_coffee.drop_duplicates(subset=['지점명'])
```
